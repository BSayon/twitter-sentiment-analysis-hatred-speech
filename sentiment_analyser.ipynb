```python
import numpy as np
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer 
```

```python
df = pd.read_csv('train.csv', index_col = 'id')
```

```python
stopwords = set(w.rstrip() for w in open('stopwords.txt'))
```

```python
wordnetlem = WordNetLemmatizer()
```

```python
def my_tokenizer(s):
    s = s.lower()
    tokens = nltk.tokenize.word_tokenize(s)
    tokens = [t for t in tokens if len(t)>2 and t.isalnum()]
    tokens = [wordnetlem.lemmatize(t) for t in tokens]
    tokens = [t for t in tokens if t not in stopwords]
    return tokens
```

```python
df['tweet'] = df['tweet'].apply(lambda x: my_tokenizer(x))
```

```python
positive_tweets = []
negative_tweets = []
word_index_map = {}
index = 0
idx = 0
for idx in range(len(traindata)):
    label = traindata.iloc[idx,0]
    tweet = traindata.iloc[idx,1]
    if label==0:
        positive_tweets.append(tweet)
    else:
        negative_tweets.append(tweet)
    for token in tweet:
        if token not in word_index_map:
            word_index_map[token] = index
            index += 1
```

```python
def tokens_to_vector(tokens, label):
    x = np.zeros(len(word_index_map) + 1)
    for t in tokens:
        i = word_index_map[t]
        x[i] += 1
    x = x/ x.sum()
    x[-1] = label
    return x
```

```python
data = np.zeros((len(traindata),len(word_index_map) + 1))
i = 0
idx = 0
for idx in range(len(traindata)):
    tweet = traindata.iloc[idx,1]
    label = traindata.iloc[idx,0]
    xy = tokens_to_vector(tweet, label)
    data[i,:] = xy
    i += 1
```
